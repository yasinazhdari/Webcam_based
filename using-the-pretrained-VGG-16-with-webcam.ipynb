```python
# Next, we will import required packages and libraries.
from tensorflow import keras
from keras.applications.vgg16  import VGG16
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input, decode_predictions
import numpy as np
```

```python
# Here, we are using MobileNet_V2 with its corresponding pre-trained weights on Image-net database as our model. use VPN in Iran!
model = VGG16(weights='imagenet')
```

```python
# Next, we'll define a function to perform prediction, and corresponding pre-processings from the MobileNet_V2 itself (not of our will)
def inference(x0):
    x1=np.copy(x0)
    x = np.expand_dims(x1, axis=0)
    x =  preprocess_input(x)
    preds = model.predict(x)
    return decode_predictions(preds, top=1)[0][0][1]
```

```python
# Here, before the main show we wnt to just test that every thing is working properly till now,
# by loading an elephant image from my computer's desktop (obviously you must change the path correspondingly)
img_path = 'C:/Users/ASUS/Desktop/elephant.jpg'
img = keras.utils.load_img(img_path, target_size=(224, 224))
x   = keras.utils.img_to_array(img)
```

```python
inference(x)
```

## In the following section, we want to use the previous Pre-trained network along with the Images which are real-time captured with the webcam. 

```python
import cv2
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    frame = cv2.resize(frame, (224, 224))
    predicted = inference(frame[...,::-1]) # BGR ==> RGB
    cv2.putText(frame,  predicted, (30, 30), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 0, 255), lineType=cv2.LINE_AA)
    cv2.imshow('Webcam', frame)
    if cv2.waitKey(1) == 13: #13 is the Enter Key
        break
cap.release()
cv2.destroyAllWindows()     
```
